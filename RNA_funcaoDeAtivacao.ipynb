# Exemplo de Função degrau

* Usado em problemas linearmente separáveis

$ \LARGE
f(x) = \begin{cases}
0 & \text{se } x < 1 \\
1 & \text{se } x \geq 1
\end{cases}
$

```python
def stepfunction(soma):
    if soma >=1:
        return 1
    return 0

step = stepfunction(-1)
print(step)
```

```python
# Criando uma lista de valores x
x_values = np.linspace(-1, 2, 100)

# Aplicando a função degrau aos valores de x
y_values = [stepfunction(x) for x in x_values]

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Função Degrau', color='black', linewidth=2)
plt.title('Gráfico da Função Degrau')
plt.xlabel('x')
plt.ylabel('H(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# Função Sigmoid

* Torna útil para problemas de classificação binária



$\LARGE
y = \frac{1}{1 + e^{-x}}
$


```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = 0.5
result = sigmoid(x)

print(f'Resultado da função sigmoide para x = 0.5: {result}')
```

```python
# Criando uma lista de valores x no intervalo [0, 1]
x_values = np.linspace(0, 1, 100)

# Aplicando a função sigmoide aos valores de x
y_values = sigmoid(x_values * 10 - 5)  # Ajustando a escala para ver a transição completa de 0 a 1

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Função Sigmoide', color='black', linewidth=2)
plt.title('Gráfico da Função Sigmoide')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

# Função Tangente Hiperbólica

* É comumente usada como função de ativação em camadas ocultas de redes neurais, especialmente em modelos recorrentes (RNNs) e em certos tipos de arquiteturas de redes neurais convolucionais (CNNs)

$
\LARGE
Y = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$

```python
# Definindo a função tangente hiperbólica
def tanh(x):
    return np.tanh(x)


x = 0.5
result = tanh(x)

print(f'Resultado da função sigmoide para x = 0.5: {result}')
```

```python
# Criando uma lista de valores x
x_values = np.linspace(-5, 5, 100)  # Gerando 100 valores entre -5 e 5

# Calculando os valores correspondentes da função tangente hiperbólica
y_values = tanh(x_values)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Função Tangente Hiperbólica', color='black', linewidth=2)
plt.title('Gráfico da Função Tangente Hiperbólica')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# ReLU - Rectified Linear Units

* Durante a fase de inferência, a função ReLU é muito eficiente computacionalmente, pois envolve apenas comparações simples e operações de atribuição

$ \LARGE
{ReLU}(x) = \max(0, x)
$

```python
def relu(x):
    return np.maximum(0, x)

# Exemplo de aplicação da função ReLU
input_value = 2.5
output_value = relu(input_value)

print("Input:", input_value)
print("Output:", output_value)
```

```python
# Criando uma lista de valores x
x_values = np.linspace(-5, 5, 100)

# Calculando os valores correspondentes da função ReLU
y_values = relu(x_values)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='ReLU', color='black', linewidth=2)
plt.title('Gráfico da Função ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# Função Linear

* Serve principalmente como uma camada de identidade em uma rede neural. Ela mantém a saída da camada anterior sem fazer nenhuma modificação, o que pode ser útil em certos casos, como em problemas de regressão simples

$\LARGE
f(x)=ax+b
$

```python
def linear(x, a, b):
    return a * x + b

# Calculando o resultado da função linear para x = 0.5
x = 0.5
a = 2
b = 1
result = linear(x, a, b)

print(f'Resultado da função linear para x = 0.5: {result}')
```

```python
# Definindo os valores de x
x_values = np.linspace(-5, 5, 100)

# Calculando os valores correspondentes da função de ativação linear
y_values = linear(x_values, a, b)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Função de Ativação Linear', color='black', linewidth=2)
plt.title('Gráfico da Função de Ativação Linear')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# Função Softmax

* Ela é usada na camada de saída da rede neural para calcular as probabilidades de pertencimento de cada classe para uma determinada entrada

$\LARGE
Softmax(z_i) = e^{z_i} / \sum_{j=1}^{K} e^{z_j}
$

```python
def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x)

# Vetor de entrada
x = np.array([2.0, 1.0, 0.1])

# Calculando as probabilidades usando a função Softmax
probabilidades = softmax(x)

print("Probabilidades de pertencimento a cada classe:")
for i, p in enumerate(probabilidades):
    print(f"Classe {i+1}: {p:.4f}")
```
